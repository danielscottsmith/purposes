***This notebook takes in R-produced data frame of STM word-by-topic probabilities, restructures them into a wordcloud-friendly format, and uses the python package wordcloud to visualize the most probable words associated with the state schooling topic.***

```python
import pandas as pd
import matplotlib.colors as mcolors
# from wordcloud import WordCloud, STOPWORDS
import os
```

```python
os.getcwd()
```

```python
ktopics = 200
```

```python
DIRECTORY = os.path.dirname(os.getcwd()) + '/'
print(DIRECTORY)
STM = DIRECTORY + '00_data/98_stm{}/'.format(ktopics)
print(STM)
VISUALS = DIRECTORY + '02_visuals/01_appendix/'
print(VISUALS)
```

```python
top_terms = pd.read_csv(STM+"top_term_probs_b.csv")
```

```python
top_terms
```

```python
list(zip(top_terms["stm200.vocab"], top_terms["X1"]))
```

```python
topics = []
i = 0 
for topic in range(1,201):
    top_terms.sort_values(by=["X{}".format(topic)], ascending=False, inplace=True)
    topics.append((i, list(zip(top_terms["stm200.vocab"], 
                               top_terms["X{}".format(topic)]))))
    i += 1
```

```python
topics[9]
```

```python
def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(0, 0)
```

```python
cloud = WordCloud(stopwords=None,
                  background_color='white',
                  width=2500,
                  height=2500,
                  max_words=100,
                  colormap='tab10',
                  color_func= grey_color_func, 
#                   color_func= lambda *args, **kwargs: colors[color_n],
                  prefer_horizontal=1, 
                 max_font_size=500)
```

```python
import random
import matplotlib.pyplot as plt
```

```python
selection = [9] 
color_n = -1
for selected in selection: 
    if color_n == 9:
        color_n = -1
    color_n += 1
    topic_words = dict(topics[selected][1])
    cloud.generate_from_frequencies(topic_words)
    plt.figure( figsize=(3,3), dpi=1000)
    plt.gca().imshow(cloud)
#     plt.gca().set_title(str(selected) + "\n", fontdict=dict(size=40))
    plt.gca().axis('off')
    plt.axis('off')
    plt.margins(x=0, y=0)
    plt.tight_layout()
    plt.savefig(VISUALS+"topic{}_stm".format(selected))
    plt.close('all')
#     fig = Image.open(VISUALS+"topic{}.jpg".format(selected))
#     os.remove(VISUALS+"topic{}.jpg".format(selected))
#     fig.save(VISUALS+"topic{}.tif".format(selected))
#     fig.close()
```
