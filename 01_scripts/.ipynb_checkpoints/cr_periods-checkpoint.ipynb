Finally, the notebook computes the degree centrality of the schooling topic in the sematic network for each year of parliamentary debate.***

### Visualize network for each period

```python
labels = pd.read_csv(LDA+"topic_labels.csv", usecols=[0,1], names=['id', 'label'], header=0)
label_dict = {labels['id'][i]: labels['label'][i]  for i in range(len(labels))} 
```

```python
def draw_ego_network(thetas_df, period, label_dict):
    """
    Builds a network visualization from a thetas dataframe based on plotly, networkx, and NetworkViz packages.
    
    Sources: 
    https://api-lab.dimensions.ai/cookbooks/8-organizations/3-Organizations-Collaboration-Network.html
    """
    
    # Intilization statement
    print("Building network {} ...".format(period))
    
    #
    # Image directory 
    #
#     clouds = os.getcwd()
#     clouds += "/clouds/" 
    cloud_dir = LDA+"01_visuals/"
    
    #
    # Colors
    #
    colors = [color for name, color in mpl.colors.XKCD_COLORS.items()]
    
    
    #
    # set up networkx
    #
    nx_g = nx.Graph()
    
    #
    # Load speeches only for the current period
    #
    coherent_topics = [k for k, v in label_dict.items() if k != v]
    period_df = thetas_df[thetas_df['period'] == period].loc[::, coherent_topics].astype(float)
    
    #
    # Dichotomize thetas by threshold
    #
    thresh = .1 #period_df.mean().mean() + 2*period_df.mean().std()
#     print(thresh)
    period_binary_df = period_df
    period_binary_df = period_binary_df.where(period_df > thresh, 0)
    period_binary_df = period_binary_df.where(period_df < thresh, 1)

    #
    # Create adjacency matrix
    #
    period_adj = period_binary_df.T.dot(period_binary_df)
    period_adj.replace(0, np.nan)
    period_adj = normalize(period_adj, norm='max')
    period_adj = pd.DataFrame(period_adj, 
                              columns=period_binary_df.columns.to_list(), 
                              index=period_binary_df.columns.to_list())
    
    #
    # create edge list
    #
    period_edg = period_adj
    period_edg.values[tuple([np.arange(len(period_adj))]*2)] = np.nan
    period_edg = period_edg.stack().reset_index()
    period_edg = pd.DataFrame(period_edg.values, columns=['to', 'fro', 'weight'])
    period_edg = period_edg[period_edg.weight > 0]
    period_edg = period_edg[period_edg.weight > 
                            (period_edg['weight'].astype(float).describe()['mean'] + 
                            period_edg['weight'].astype(float).describe()['std']*0)]

    #
    # add nodes to graph
    #
    nodes = set(list(period_edg['to']) + list(period_edg['fro']))  
    for node in nodes:
        nx_g.add_node(
        node,
        label = label_dict[node],
        title = f"<h1>{label_dict[node]}</h1>",
        value = period_df[node].describe()['mean'] * period_df[node].describe()['mean'], 
        borderWidthSelected = 5,
        shape = "dot",
        )

    #
    # add weighted edges
    #
    for index, row in period_edg.iterrows():
        nx_g.add_edge(row['fro'], row['to'],
            value = float(row['weight']),
            arrows = "none")
         
    #
    # find modules using Blondel's Multilevel algorithm 
    #
    ig_g = igraph.Graph.from_networkx(nx_g)
    ig_nx_concord = dict(zip(list(range(len(nx_g.nodes))), nx_g.nodes))
    hoods = ig_g.community_multilevel()
    hood_colors = {}
    for hood in hoods:
        random_color = random.sample(sorted(colors), 10)[3]
        while random_color in hood_colors.items():
            random_color = random.sample(sorted(colors), 10)[8]
        for node in hood: 
            if node not in hood_colors:
                hood_colors[ig_nx_concord[node]] = random_color
                
    nx.set_node_attributes(nx_g, hood_colors, "color")
    
    schooling = nx.ego_graph(nx_g, 'topic106', radius=1, center=True, undirected=True, distance=None)
    
    # 
    # set up dataviz
    #
    g = NetworkViz(notebook=False, width="80%", height="100%")
    g.from_nx(schooling)
    g.force_atlas_2based(gravity=-30, 
                         central_gravity=0.001, 
                         spring_length=110, 
                         spring_strength=0.2, 
                         damping=0.9, 
                         overlap=0.02) 
                        #gravity=-50, central_gravity=0.001, spring_length=50, spring_strength=0.04, damping=0.4, overlap=0
    g.toggle_hide_edges_on_drag(False)
    g.set_edge_smooth("straightCross")
    g.show_buttons(True)

    #
    # add tooltips with neighbor info
    #
    neighbor_map = g.get_adj_list()
    for node in g.nodes:
        neighbors = neighbor_map[node["id"]]
        ls = [neighbor for neighbor in neighbors]
        labels = []
        for label in ls:
            labels.append(label_dict[label])
        node["title"] += "<img src='{}{}.jpg'><br><br>".format(cloud_dir, node['id']) + "<h4>{} appears in speeches with:</h4><ol><li>".format(node["label"], len(labels)) + "</li><li>".join(labels)
    g.write_html(VISUALS+"00_text/"+"{}.html".format(period))
    print("... complete!")
    print('\n\n')
    return schooling
```

```python
for period in thetas_periods.period.unique():
    draw_ego_network(thetas_periods, period, label_dict)
```

## Measures of centrality in semantic networks each year

```python
yr_eigen = []
yr_degree = []
for year, yr_df in year_speeches:
    row_eigen = [year, yr_df.shape[0]]
    row_degree = [year, yr_df.shape[0]]
    
    # if STM:
#     yr_thetas = yr_df.loc[::, ['X' + str(i) for i in range(1,ktopics+1)]].astype(float)
    
    # IF LDA
    yr_thetas = yr_df.loc[::, ['topic' + str(i) for i in range(ktopics)]].astype(float)
    
    #
    # Dichotomize theras
    #
    thresh = .05
    yr_thetas.where(yr_thetas > thresh, 0, inplace=True)
    yr_thetas.where(yr_thetas < thresh, 1, inplace=True)
     
    #
    # Create adjacency matrix & network
    #
    yr_adj = normalize(yr_thetas.T.dot(yr_thetas))
    n_yr_adj = pd.DataFrame(yr_adj)
    network = nx.from_pandas_adjacency(n_yr_adj)
    
    # compute eigenvector centrality
    eigen = list(nx.eigenvector_centrality(network, max_iter=600).values())
    row_eigen.extend(eigen)
    yr_eigen.append(row_eigen)
    
    # compute degree centrality 
    degree = list(nx.degree_centrality(network).values())
    row_degree.extend(degree)
    yr_degree.append(row_degree)

#
# Write stata file of degree 
#
dcols = ['year', 'nyr']
ecols = ['year', 'nyr']
dcols.extend(['dtopic' + str(i) for i in range(ktopics)]) # WARNING! STM TOPICS WILL BE INDEX-1
ecols.extend(['etopic' + str(i) for i in range(ktopics)]) # WARNING! STM TOPICS WILL BE INDEX-1
yr_eigen_df = pd.DataFrame(yr_eigen, columns=ecols)
yr_eigen_df.to_stata(DVS+'eigen_lda.dta', write_index=False) 
# yr_eigen_df.to_stata(DVS+'eigen_stm.dta', write_index=False) 
yr_degree_df = pd.DataFrame(yr_degree, columns=dcols)
yr_degree_df.to_stata(DVS+'degree_lda.dta', write_index=False) 
# yr_degree_df.to_stata(DVS+'degree_stm.dta', write_index=False) 
```
